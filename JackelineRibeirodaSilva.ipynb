````python
def gera_mapa():
  mapa = []
  mapa.append(list('XXXXXXXXXXXXX'))
  mapa.append(list('X           X'))
  mapa.append(list('X   X  XX   X'))
  mapa.append(list('X   X  XX   X'))
  mapa.append(list('X           X'))
  mapa.append(list('XX XXXXXXX XX'))
  mapa.append(list('X @ X       X'))
  mapa.append(list('X   X  XX   X'))
  mapa.append(list('X   X  XX   X'))
  mapa.append(list('XXXXXXXXXXXXX'))
  return mapa

mapa = gera_mapa()
for l in range(len(mapa)):
  for c in range(len(mapa[l])):
    print(mapa[l][c],end="")
    if c == len(mapa[l])-1:
       print()

import random

def espaco_vazio(mapa):
  vazio = 0
  for l in range(0,len(mapa)):
    for c in range(0,len(mapa[0])):
      if mapa[l][c] == ' ':
        vazio = vazio + 1
  return vazio    

def posicao_robo(mapa):
  for l in range(0,len(mapa)):
    for c in range(0,len(mapa[0])):
      if mapa[l][c] == '@':
        return l,c

def limpa_casa (mapa):
  posicao = [ ]
  mapa = gera_mapa()
  N = 0
  S = 1
  L = 2
  O = 3
  while espaco_vazio (mapa) != 0 :
    sorte = random.randint(0,3)
    linha_robo, coluna_robo = posicao_robo(mapa) 
    if  sorte == 0:
      if mapa[linha_robo-1][coluna_robo] == ' ' or mapa[linha_robo-1][coluna_robo] =='-':
         mapa[linha_robo-1][coluna_robo] = '@'
         mapa[linha_robo][coluna_robo] = '-'
         posicao.append('N')
    elif  sorte == 1:
      if mapa[linha_robo + 1][coluna_robo] == ' ' or mapa[linha_robo +1][coluna_robo] =='-':
         mapa[linha_robo+1][coluna_robo] = '@'
         mapa[linha_robo][coluna_robo] = '-'
         posicao.append('S')
    elif  sorte == 2:
      if mapa[linha_robo ][coluna_robo + 1] == ' ' or mapa[linha_robo][coluna_robo + 1] =='-':
         mapa[linha_robo][coluna_robo + 1] = '@'
         mapa[linha_robo][coluna_robo] = '-'
         posicao.append('L')
    elif  sorte == 3:
      if mapa[linha_robo][coluna_robo-1] == ' ' or mapa[linha_robo][coluna_robo-1] =='-':
         mapa[linha_robo][coluna_robo-1] = '@'
         mapa[linha_robo][coluna_robo] = '-'
         posicao.append('O')  
  return posicao

posicao = limpa_casa(mapa)
for i in range(0,len(posicao)):
   print(posicao[i],end='')        
````
